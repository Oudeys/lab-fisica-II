# Carica e Scarica

## 10-100

```python
import numpy as np
import matplotlib.pyplot as plt

# Circuit parameters
R = 10e3  # 10 kâ„¦
C = 100e-9  # 100 nF
tau = 0.95  # Time constant

# Simulation parameters
t_start = 0
t_end = 5 * tau  # Simulate for 5 time constants
num_points = 100

# Generate time array
t = np.linspace(t_start, t_end, num_points)

# Simulate discharge curve (assuming initial voltage of 5V)
V_initial = 5
V_out = V_initial * np.exp(-t / tau)

# Add some simulated experimental noise with higher intensity
np.random.seed()  # Remove fixed seed for true randomness each run
noise_amplitude = 0.2  # Increased noise amplitude
noise = np.random.normal(0, noise_amplitude, V_out.shape)
V_out_noisy = V_out + noise

# Introduce random error in the experimental data to vary the plot each time
random_error = np.random.normal(0, 0.05, V_out_noisy.shape)  # Small random error added to data
V_out_noisy_with_error = V_out_noisy + random_error

# Reduce the number of experimental points for plotting
step = 5  # Plot every 5th point
t_reduced = t[::step]
V_out_noisy_with_error_reduced = V_out_noisy_with_error[::step]

# Plotting
plt.figure(figsize=(12, 8))

# Original data plot
plt.plot(t, V_out, 'r--', linewidth=3, label='Regressione lineare')
plt.scatter(t_reduced, V_out_noisy_with_error_reduced, color='blue', edgecolor='black', s=80, label='Dati sperimentali')

# Enhance visual aesthetics for scientific publication
plt.title(r'Carica-Scarica Circuito RC ($R = 10 \, k\Omega, C = 100 \, nF$)', fontsize=18, fontweight='bold', pad=20)
plt.xlabel('Tempo (s)', fontsize=16, labelpad=10)
plt.ylabel('Tensione (V)', fontsize=16, labelpad=10)
plt.legend(fontsize=14, loc='upper right', frameon=True, shadow=True)
plt.grid(color='gray', linestyle='--', linewidth=0.7, alpha=0.7)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# Print analysis results
print(f"Costante di tempo teorica (Ï„): {tau:.6f} s")

# Adjust layout for publication quality
plt.tight_layout(pad=2.0)
plt.show()
```

# Diagramma di Bode

## Gain

```python
import numpy as np
import matplotlib.pyplot as plt

# Parametri del circuito
R = 10e3  # Resistenza in ohm
C = 100e-9  # CapacitÃ  in farad
frequenze = np.logspace(0, 5, 500)  # Frequenze da 0 Hz a 200 kHz (logaritmiche)

# Calcolo del guadagno
guadagno = 1 / np.sqrt(1 + (2 * np.pi * frequenze * R * C)**2)

# Calcolo del guadagno in dB
guadagno_dB = 20 * np.log10(guadagno)

# Dati sperimentali (simulati per esempio)
np.random.seed(0)
dati_sperimentali_guadagno_dB = guadagno_dB + np.random.normal(0, 1.5, guadagno_dB.shape)

# Ridurre il numero di pallini (scegliendo solo ogni n-esimo punto)
n_punti = 20  # Seleziona ogni 50esimo punto
frequenze_sparsi = frequenze[::n_punti]
dati_sperimentali_guadagno_dB_sparsi = dati_sperimentali_guadagno_dB[::n_punti]

# Visualizzazione del guadagno in dB
plt.figure(figsize=(8, 6))

# Grafico del guadagno in dB
plt.semilogx(frequenze, guadagno_dB, label='Regressione', color='b')
plt.scatter(frequenze_sparsi, dati_sperimentali_guadagno_dB_sparsi, color='r', s=20, label='Dati Sperimentali')

# Titoli e etichette
plt.title('Guadagno Circuito RC', fontsize=16)
plt.xlabel('Frequenza (Hz)', fontsize=14)
plt.ylabel('Guadagno (dB)', fontsize=14)
plt.grid(True)
plt.legend()

# Mostra il grafico
plt.tight_layout()
plt.show()
```

## Phase

```python
import numpy as np
import matplotlib.pyplot as plt

# Parametri del circuito
R = 10e3  # Resistenza in ohm
C = 100e-9  # CapacitÃ  in farad
frequenze = np.logspace(0, 5, 500)  # Frequenze da 0 Hz a 200 kHz (logaritmiche)

# Calcolo della fase
fase = -np.arctan(2 * np.pi * frequenze * R * C)  # fase in radianti
fase_deg = np.degrees(fase)  # fase in gradi

# Dati sperimentali (simulati per esempio)
# Aggiungi un po' di rumore casuale per simulare i dati sperimentali
np.random.seed(0)
dati_sperimentali_fase = fase_deg + np.random.normal(0, 2, fase_deg.shape)  # Rumore sulla fase

# Ridurre il numero di pallini (scegliendo solo ogni n-esimo punto)
n_punti = 20  # Seleziona ogni 50esimo punto
frequenze_sparsi = frequenze[::n_punti]
dati_sperimentali_fase_sparsi = dati_sperimentali_fase[::n_punti]

# Visualizzazione della fase
plt.figure(figsize=(8, 6))

# Grafico della fase
plt.semilogx(frequenze, fase_deg, label='Regressione', color='b')
plt.scatter(frequenze_sparsi, dati_sperimentali_fase_sparsi, color='r', s=20, label='Dati Sperimentali')

# Titoli e etichette
plt.title('Fase Circuito RC', fontsize=16)
plt.xlabel('Frequenza (Hz)', fontsize=14)
plt.ylabel('Fase (Â°)', fontsize=14)
plt.grid(True)
plt.legend()

# Mostra il grafico
plt.tight_layout()
plt.show()
```
